```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/SBAnational.csv')
pd.options.display.max_columns = 999
df
```

```python
#州先只選CA
df_CA = df[df['State'] == 'CA'].copy().reset_index(drop = True)
```

```python
#處理Datetime
date_columns = ['ApprovalDate', 'ChgOffDate','DisbursementDate']
for col in date_columns:
    df_CA[col] = pd.to_datetime(df_CA[col], format = '%d-%b-%y')
```

```python
#處理Currency
amount_columns = ['DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv']
for col in amount_columns:
    df_CA[col] = df_CA[col].replace('[\$,]', '', regex=True).astype(float) 
```

```python
#空值
df_CA.isnull().sum()
```

```python
#把空值的row先刪掉
df_CA.dropna(axis = 0, subset = ['DisbursementDate'])
df_CA.reset_index(drop=True)
```

```python
df_CA.groupby(df_CA['DisbursementDate'].map(lambda x: x.year))['DisbursementGross'].sum()/1000000
```
